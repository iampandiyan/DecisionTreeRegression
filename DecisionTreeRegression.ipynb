```python
import pandas as pd
```

```python
dataset=pd.read_csv("50_Startups.csv")
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['R&D Spend', 'Administration', 'Marketing Spend', 'State_Florida', 'State_New York']]
independent
```

```python
dependent=dataset[['Profit']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
print(X_train)
print(Y_train)
print(X_test)
print(Y_test)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
X_train
X_test
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor=DecisionTreeRegressor(criterion='absolute_error',max_features=None,splitter='best')
regressor=regressor.fit(X_train,Y_train)
```

```python
import matplotlib.pyplot as plt
from sklearn import tree
tree.plot_tree(regressor)
plt.show()
```

```python
Y_predict=regressor.predict(X_test)
Y_predict
```

```python
from sklearn.metrics import r2_score
r2=r2_score(Y_test,Y_predict)
r2
```

```python
import pickle
pickle.dump(regressor,open('finalized_DesisionTree_model.sav','wb'))
```

```python
loaded_model=pickle.load(open('finalized_DesisionTree_model.sav','rb'))
preinput=sc.transform([[1234,345,4565,1,0]])
result=loaded_model.predict(preinput)
print(preinput)
print(result)
```
